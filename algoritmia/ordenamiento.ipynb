# Algoritmia en Python — Ordenamiento y Búsqueda

En este cuaderno vamos a implementar y comparar 3 algoritmos con `timeit.repeat()` + `min()`
como aprendimos en `time.ipynb`. La idea es **ver** la diferencia entre complejidades
(O(n), O(log n), O(n²)) en práctica, no solo en teoría.

## Los 3 algoritmos

| # | Algoritmo | Complejidad | Idea |
|---|-----------|------------|------|
| 1 | Burbuja | O(n²) | Compara adyacentes y los intercambia si están mal ordenados |
| 2 | Búsqueda lineal | O(n) | Recorre uno por uno hasta encontrar |
| 3 | Búsqueda binaria | O(log n) | Divide la lista a la mitad cada vez (requiere lista ORDENADA) |

```python
#importamos lo necesario para medir y los algoritmos de ordenamiento
import timeit
import random
from typing import Callable
n = 1000

l = []
for i in range(n):
    l.append(round(i+1*random.uniform(5,100)))

print(f"Lista sin ordenar {l}")

print("Importe hecho correctamente, lista creada correctamente.")
```

# Benchmark

Usaremos la siguientefuncion para definir cuanto tarda una funcion en tiempo de ejecion usando timeit.request, con un numero de 10000 ejeciciones, 5 repeticiones, escogiendo el resultado de menor valor.

```python
def benchmark(fun : Callable, name : str):
    t = timeit.Timer(fun)
    n, _ = t.autorange()
    arr = t.repeat(repeat=5, number=n)
    return print(f"La funcion {name}, tardo {min(arr)}")
```

# Algoritmo Burbuja

Este algotimo es un agoritmo de comparacion, que lo que hace es recorrer toda una lista, varias veces, con 2 indixes, uno mayor al otro por 1, es decir un i y i+1, comparando que siempre i sea menor a i+a, si esto no es asi, los cambia de lugar y continua, asi multiples veces hasta ordenar toda la lista.

```python
def burbuje_algoritm(arr) -> list:
    arr_burbuja = list(arr)
    n_arr = len(arr_burbuja)
    for i in range(n_arr-1):
        for j in range(n_arr - i -1):
            if arr_burbuja[j] > arr_burbuja[j+1]: # 3 - 2
                aux = arr_burbuja[j+1] # 2
                arr_burbuja[j+1] = arr_burbuja[j] # j+1 = 2 -> 3
                arr_burbuja[j] = aux # j = 3 -> 2
                # j = 2 j+1 = 3
    return arr_burbuja             
    
#Ejecutamos    
print(f"lista ordenada {burbuje_algoritm(l)}")    

benchmark(lambda : burbuje_algoritm(l), f"algoritmo de ordenamiento burbuja con {n} elementos")    
```
