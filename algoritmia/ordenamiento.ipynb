# Benchmark

Usaremos la siguientefuncion para definir cuanto tarda una funcion en tiempo de ejecion usando timeit.request, con un numero de 10000 ejeciciones, 5 repeticiones, escogiendo el resultado de menor valor.

```python
def benchmark(fun : Callable, name : str):
    t = timeit.Timer(fun)
    n, _ = t.autorange()
    arr = t.repeat(repeat=5, number=n)
    return print(f"La funcion {name}, tardo {min(arr)}")
```

# Algoritmo Burbuja

Este algotimo es un agoritmo de comparacion, que lo que hace es recorrer toda una lista, varias veces, con 2 indixes, uno mayor al otro por 1, es decir un i y i+1, comparando que siempre i sea menor a i+a, si esto no es asi, los cambia de lugar y continua, asi multiples veces hasta ordenar toda la lista.

```python
def burbuje_algoritm(arr) -> list:
    arr_burbuja = list(arr)
    n_arr = len(arr_burbuja)
    for i in range(n_arr-1):
        for j in range(n_arr - i -1):
            if arr_burbuja[j] > arr_burbuja[j+1]: # 3 - 2
                aux = arr_burbuja[j+1] # 2
                arr_burbuja[j+1] = arr_burbuja[j] # j+1 = 2 -> 3
                arr_burbuja[j] = aux # j = 3 -> 2
                # j = 2 j+1 = 3
    return arr_burbuja             
    
#Ejecutamos    
arr_burbuja = burbuje_algoritm(l)
print(f"lista ordenada {arr_burbuja}")    

benchmark(lambda : burbuje_algoritm(l), f"algoritmo de ordenamiento burbuja con {n} elementos")    
```

# QuickShort

Quickshort es un algoritmo que se basa en selecionar un numero de forma aleatoria en en una lista que se quiera ordenar, este elemento se le llama pivot, y es cuyo elemtnos se usara para ordenar la lista, el pivot puede cambiar su valor solo si su lista queda ordenada, es decir.

Cuando se seleciona un pivot, se valida el resto de elemtnso contra el, para los elementos mayores al pivot, se los pone al lado derecho, paralos menores al izquierdo, asi generando listas, en cada lista se hace lo mismo recursivamente.

```python
def quickshort(arr):
    arr_l = list(arr)
    _quickshort(arr_l, 0, len(arr_l)-1)

def _quickshort(arr, low, high):
    if low >= high:
        return
    pivot = parttion(arr, low, high)
    _quickshort(arr, low, pivot-1)
    _quickshort(arr, pivot+1, high)

def parttion(arr, low, high):
    pivot = arr[high]
    i = low
    for j in range(low, high):
        if arr[j] <= pivot:
            arr[j], arr[i] = arr[i], arr[j]
            i += 1
    arr[i], arr[high] = arr[high], arr[i]
    return i
    
quickshort(l)
benchmark(lambda : quickshort(l), f"Algoritmo quickshort con {n} numero de elementos")
        
```
