# Algoritmo Burbuja

Este algotimo es un agoritmo de comparacion, que lo que hace es recorrer toda una lista, varias veces, con 2 indixes, uno mayor al otro por 1, es decir un i y i+1, comparando que siempre i sea menor a i+a, si esto no es asi, los cambia de lugar y continua, asi multiples veces hasta ordenar toda la lista.

```python
n = 10
l = []
for i in range(n):
    l.append(i+1*random.uniform(5,100))

print(f"Lista sin ordenar {l}")

def  burbuje_algoritm(arr) -> list:
    n_arr = len(arr)
    for i in range(n_arr-1):
        for j in range(n_arr - i -1):
            if arr[j] > arr[j+1]: # 3 - 2
                aux = arr[j+1] # 2
                arr[j+1] = arr[j] # j+1 = 2 -> 3
                arr[j] = aux # j = 3 -> 2
                # j = 2 j+1 = 3
    return arr             
    
#Ejecutamos    
print(f"lista {burbuje_algoritm(l)}")        
```
